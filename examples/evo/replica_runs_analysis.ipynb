```python
import pandas as pd
loaded_files = []
prefix = "obi-wan"
num_files = 5
headers = ["{0}{1}".format(prefix, i) for i in range(num_files)]
for i in range(num_files):
    moo = pd.read_csv("./runs/{0}{1}".format(prefix, i))
    moo['replicant'] = "{0}{1}".format(prefix, i)
    moo.set_index(["replicant"])
    loaded_files.append(moo)

loaded_files[0]

squish = pd.concat(loaded_files, axis=0)
#squish
squish = squish.set_index(["replicant", "Update"])
squish.loc["obi-wan1"]
#squish.loc["{}0".format(prefix)]
```

Normal distribution of 0th graph
--------------------------------

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

graphnum = 1

sns.set(style="white", palette="muted", color_codes=True)
sns.distplot(squish.loc["obi-wan1".format(prefix, graphnum)]["ShannonDiversity"], bins=10)
```

Shannon Diversity of 0th Cell

```python
plt.plot(squish.loc["{0}1".format(prefix, graphnum)]["ShannonDiversity"])
plt.show()
```
